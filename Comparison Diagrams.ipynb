```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
healthy_df = pd.read_csv('healthy.csv', delimiter=",")
                         
```

```python
parkinsons_df = pd.read_csv('parkinsons.csv', delimiter=",")
```

```python
n = 1  # This is our first dataset (out of 2)
t = 2 # Number of datasets
d = len(['jitter_local','jitter_local_absolute','jitter_rap','jitter_ppq5','jitter_ddp']) # Number of sets of bars
w = 0.8 # Width of each bar
healthy_ax = [t*element + w*n for element
             in range(d)]
```

```python
n = 2  # This is our first dataset (out of 2)
t = 2 # Number of datasets
d = len(['jitter_local','jitter_local_absolute','jitter_rap','jitter_ppq5','jitter_ddp']) # Number of sets of bars
w = 0.8 # Width of each bar
parkinsons_ax = [t*element + w*n for element
             in range(d)]
```

```python
sns.kdeplot(healthy_df['jitter_local'], shade=True, label="Healthy Individuals", color='#1228EE')
sns.kdeplot(parkinsons_df['jitter_local'], shade=True, label="Parkinson's Individuals", color='#3DEB1A')
plt.title('Jitter Measure 1')
sns.set_style("whitegrid")
plt.savefig('jitter_measure_1.png')
plt.legend()
plt.show()
```
